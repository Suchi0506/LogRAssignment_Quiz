```python
df= pd.read_csv('census-income .csv')
```

```python
df.head()
```

```python
#Replacing "?" with NaN in 'occupation'
df['occupation']= df['occupation'].replace('?', np.nan)

# counting number of occupations
count= df['occupation'].nunique()
print(count)
```

```python
# Counting the people working as tech support and have an annual income greater than 50k
count = df[(df['occupation'] == 'Tech-support') & (df['annual_income'] == '>50K')]

# Getting the number of such people
print(f"Number of people working as Tech-support with an income >50K: {len(count)}")
```

```python
# Total number of missing values
df.replace('?', np.nan, inplace=True)

# Counting total missing values in the dataset
total_missing_values = df.isnull().sum().sum()

total_missing_values
```

```python
# counting number of people that are having private work classes and are not from the United States of America?
count= df[(df['workclass'] == 'Private') & ~ (df['native-country'] == 'United-States')]
len(count)
```

```python
# counting no. of people either having Annual Income less than or equal to 50k or their working hours is greater than or equal to 40 hrs:
count = df[(df['annual_income'] == '<=50K') | (df['hours-per-week'] >= 40)]

print(f"Total count: {len(count)}")
```

```python
#Renaming the last column as Annual Income
df.rename(columns={'annual_income': 'Annual Income'}, inplace=True)
```

```python
#Removing the missing values from the dataset
df.dropna(inplace=True)

df.isnull().sum()
```

```python
#Changing the labels of categorical data into numerical data using Label Encoder.
from sklearn.preprocessing import LabelEncoder
label_encoder= LabelEncoder()

# Label Encoding for binary columns
df['sex'] = label_encoder.fit_transform(df['sex'])
df['Annual Income'] = label_encoder.fit_transform(df['Annual Income'])

df.head()
```

```python
#Splitting the dataset into a train and test of proportions 70:30 and set the random state to 0.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X= df.drop('Annual Income', axis=1)
y= df['Annual Income']
```

```python
# Applying One-Hot Encoding for multi-category columns like 'workclass', 'occupation', etc.
X = pd.get_dummies(X, drop_first=True)

# Splitting the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

logreg = LogisticRegression

logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the Logistic Regression model: {accuracy * 100:.2f}%")

```

```python
#Specificity of the model
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
specificity = tn / (tn + fp)
print(f"Specificity of the model: {specificity * 100:.2f}%")
```

```python
# model’s precision when the target is False
from sklearn.metrics import precision_score
# Precision for the False class (negative class)
precision_false_class = tn / (tn + fn)
print(f"Precision (False class) of the model: {precision_false_class * 100:.2f}%")

```

```python
# Total support value from the above model

total_support = (tp + fn) + (tn + fp)
print(f"Total support: {total_support}")

```

```python
# f1 score of the above model when the target is True
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print(f"F1 score of the model: {f1 * 100:.2f}%")
```

```python
# Correctly classified records
correctly_classified = tp + tn
print(f"Correctly classified records: {correctly_classified}")
```
